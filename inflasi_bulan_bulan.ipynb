```python

import matplotlib.pyplot as plt
# Create DataFrame
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Sample data provided
df = pd.read_csv('tingkat-inflasi-tahun-ke-tahun-menurut-kabupaten-kota.csv')

df = df.loc[:, df.loc[df['tahun'] != 2013].all()]

df_updated = pd.DataFrame(df)

# Plotting the updated data
plt.figure(figsize=(12, 8))
sns.lineplot(data=df_updated, x='bulan', y='tingkat_inflasi_tahun_ke_tahun', hue='bps_nama_kabupaten_kota', marker='o')

plt.title('Tingkat Inflasi Bulan ke Bulan (2014)')
plt.xlabel('Bulan')
plt.ylabel('Tingkat Inflasi')
plt.xticks(rotation=45)
plt.legend(title='Kabupaten/Kota')
plt.tight_layout()

# Save the plot
# plt.savefig('/mnt/data/tingkat_inflasi_2014_updated.png')

plt.show()
```

```python
import plotly.express as px

# Filter the DataFrame for the year 2014
df_filtered = df_updated[df_updated['tahun'] == 2015]

# Create the line chart using the filtered data
fig = px.line(df_filtered, 
              x='bulan', 
              y='tingkat_inflasi_tahun_ke_tahun', 
              color='bps_nama_kabupaten_kota', 
              title='Tingkat Inflasi Bulan ke Bulan (2014)',
              labels={'tingkat_inflasi_bulan_ke_bulan': 'Tingkat Inflasi', 'bulan': 'Bulan'},
              markers=True)

# Display the plot
fig.show()
```

```python
import plotly.express as px

# Assuming your DataFrame is named df_filtered for the year 2014
heatmap_data = df_filtered.pivot(index='bulan', columns='bps_nama_kabupaten_kota', values='tingkat_inflasi_tahun_ke_tahun')

# Create the heatmap
fig = px.imshow(heatmap_data, 
                labels=dict(x="Kabupaten/Kota", y="Bulan", color="Tingkat Inflasi"),
                title="Heatmap of Tingkat Inflasi Bulan ke Bulan (2014)",
                aspect="auto")

# Display the plot
fig.show()
```

```python
import plotly.express as px

# Filter the DataFrame for the year 2014
df_filtered = df_updated[df_updated['tahun'] == 2014]

# Create the stacked bar chart
fig = px.bar(df_filtered, 
             x='bulan', 
             y='tingkat_inflasi_tahun_ke_tahun', 
             color='bps_nama_kabupaten_kota', 
             title='Stacked Bar Chart of Tingkat Inflasi Bulan ke Bulan (2014)',
             labels={'tingkat_inflasi_bulan_ke_bulan': 'Tingkat Inflasi', 'bulan': 'Bulan'},
             barmode='stack')

# Display the plot
fig.show()
```
